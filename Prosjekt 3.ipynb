### Exercise 2.1
We have to show that the function

$$
    \tilde{\phi}(x, t) = \frac{1}{\sqrt{4\pi D t}}e^{-\frac{(x-\mu)^2}{4D t}},
$$
where $D$ is a constant, is a solution of the diffusion equation,
$$
    \frac{\partial\phi(x, t)}{\partial t} = D\frac{\partial^2 \phi(x, t)}{\partial x^2}.
$$

Let us start with the left side of the equation:
$$
\begin{align}
    \frac{\partial\tilde{\phi}(x, t)}{\partial t}&= \
    \frac{\partial}{\partial t}\left(\frac{1}{\sqrt{4\pi D t}}e^{-\frac{(x-\mu)^2}{4D t}}\right),\\
    &= \
    \frac{\partial}{\partial t}\left(\frac{1}{\sqrt{4\pi D t}}\right) e^{-\frac{(x-\mu)^2}{4D t}} +\
    \frac{1}{\sqrt{4\pi D t}}\frac{\partial}{\partial t}\left(e^{-\frac{(x-\mu)^2}{4D t}}\right), \\
    &= \
    \left(-\frac{1}{2t\sqrt{4\pi D t}}\right)e^{-\frac{(x-\mu)^2}{4D t}} +\
    \frac{1}{\sqrt{4\pi D t}}\left(\frac{(x-\mu)^2}{4D t^2}e^{-\frac{(x-\mu)^2}{4D t}}\right),\\
    &= \
    \frac{1}{\sqrt{4\pi D t}}e^{-\frac{(x-\mu)^2}{4D t}}\
    \left(\frac{(x-\mu)^2}{4D t^2}-\frac{1}{2t}\right),\\
    &= \
    \tilde{\phi}(x, t)\left(\frac{(x-\mu)^2}{4D t^2}-\frac{1}{2t}\right).
\end{align}
$$

Now lets compute the right side:
$$
\begin{align}
    D\frac{\partial^2\tilde{\phi}(x,t)}{\partial x^2} &= \
    D\frac{\partial^2}{\partial x^2}\left(\frac{1}{\sqrt{4\pi D t}}e^{-\frac{(x-\mu)^2}{4D t}}\right),\\
    &= \
    D\frac{1}{\sqrt{4\pi D t}}\frac{\partial^2}{\partial x^2}\left(e^{-\frac{(x-\mu)^2}{4D t}}\right),\\
    &= \
    D \frac{1}{\sqrt{4\pi D t}}\frac{\partial}{\partial x}\
    \left(e^{-\frac{(x-\mu)^2}{4D t}}\left(-2\frac{x-\mu}{4D t}\right)\right), \\
    &= \
    D \frac{1}{\sqrt{4\pi D t}}\left[\frac{\partial}{\partial x}\
    \left(e^{-\frac{(x-\mu)^2}{4D t}}\right)\left(-2\frac{x-\mu}{4D t}\right) + \
    e^{-\frac{(x-\mu)^2}{4D t}}\frac{\partial}{\partial x}\left(-2\frac{x-\mu}{4D t}\right)\right],\\
    &= \
    D \frac{1}{\sqrt{4\pi D t}}e^{-\frac{(x-\mu)^2}{4D t}}\left(\
    \left(-2\frac{x-\mu}{4D t}\right)^2-\frac{2}{4D t}\right),\\
    &= \
    D \tilde{\phi}(x, t)\left(\frac{(x-\mu)^2}{4D^2 t^2}-\frac{1}{2D t}\right),\\
    &= \
    \tilde{\phi}(x, t)\left(\frac{(x-\mu)^2}{4D t^2}-\frac{1}{2t}\right).
\end{align}
$$

We see that $\tilde{\phi}(x, t)$ is indeed a solution of the diffusion equation.

#### 2.2.1
With the initial boundary $\phi(x,0) = \delta(x-x_0)$, and the Taylor expansion of $\phi(x,t)$
$$
    \phi(x,t) = \phi(x,0) + \
    \sum_{n=1}^\infty \frac{t^n}{n!}\left.\frac{\partial^n \phi(x,t')}{\partial t'^n}\right|_{t' = 0},
$$
let us show that the Taylor expression of $\phi(x,t)$ corresponds to the integral
$$
    \phi(x,t) = \frac{1}{2\pi} \int_{-\infty}^{\infty} dk e^{-Dtk^2}e^{-ik(x-x_0)}.
$$

$$
\begin{align}
    \phi(x,t) &= \phi(x,0) + \
    \sum_{n=1}^\infty \frac{t^n}{n!}\left.\frac{\partial^n \phi(x,t')}{\partial t'^n}\right|_{t' = 0},\\
    &= \
    \phi(x,0) + \sum_{n=1}^\infty \frac{(Dt)^n}{n!}\frac{\partial^{2n} \phi(x,0)}{\partial x^{2n}},\\
    &= \
    \delta(x-x_0) + \sum_{n=1}^\infty \frac{(Dt)^n}{n!}\frac{\partial^{2n} \delta(x-x_0)}{\partial x^{2n}},\\
    &= \
    \frac{1}{2\pi}\int_{-\infty}^{\infty}dk e^{-ik(x-x_0)} + \
    \sum_{n=1}^\infty \frac{(Dt)^n}{n!}\frac{\partial^{2n}}{\partial x^{2n}}\left(
    \frac{1}{2\pi}\int_{-\infty}^{\infty}dk e^{-ik(x-x_0)}\right),\\
    &= \
    \frac{1}{2\pi}\int_{-\infty}^{\infty}dk e^{-ik(x-x_0)} + \
    \frac{1}{2\pi}\sum_{n=1}^\infty \frac{(Dt)^n}{n!}
    \int_{-\infty}^{\infty}dk \frac{\partial^{2n}}{\partial x^{2n}}e^{-ik(x-x_0)},\\
    &= \
    \frac{1}{2\pi}\int_{-\infty}^{\infty}dk e^{-ik(x-x_0)} + \
    \frac{1}{2\pi}\int_{-\infty}^{\infty}\sum_{n=1}^\infty dk \frac{(Dt)^n}{n!}(-ik)^{2n} e^{-ik(x-x_0)},\\
    &= \
    \frac{1}{2\pi}\int_{-\infty}^{\infty} \sum_{n=0}^\infty dk \frac{(-Dtk^2)^n}{n!}e^{-ik(x-x_0)},\\
    &= \
    \frac{1}{2\pi} \int_{-\infty}^{\infty} dk e^{-Dtk^2}e^{-ik(x-x_0)}.
\end{align}
$$

Good! Now let's compute the integral:
$$
\begin{align}
    \phi(x,t) &= \frac{1}{2\pi} \int_{-\infty}^{\infty} dk e^{-Dtk^2}e^{-ik(x-x_0)}, \qquad \
    \text{Substituting $a = Dt \,(> 0)$, and $b = i(x-x_0)$},\\
    &= \
    \frac{1}{2\pi} \int_{-\infty}^{\infty} dk e^{-(ak^2+bk)}, \\
    &= \
    \frac{1}{2\pi} \int_{-\infty}^{\infty} \
    e^{-\left[\left(\sqrt{a}k+\frac{b}{2\sqrt{a}}\right)^2 - \frac{b^2}{4a} \right]} dk, \\
    &= \
    \frac{1}{2\pi}e^{\frac{b^2}{4a}} \int_{-\infty}^{\infty} \
    e^{-\left(\sqrt{a}k+\frac{b}{2\sqrt{a}}\right)^2} dk, \qquad \
    \text{Substituting $u = \sqrt{a}k+b\,/\,(2\sqrt{a})$},\\
    &= \
    \frac{1}{2\pi}e^{\frac{b^2}{4a}} \int_{-\infty}^{\infty} \frac{1}{\sqrt{a}}e^{-u^2} du, \\
    &= \
    \frac{1}{2\pi}e^{\frac{b^2}{4a}} \sqrt{\frac{\pi}{a}}, \\
    &= \
    \frac{1}{\sqrt{4\pi a}}e^{\frac{b^2}{4a}}, \\
    &= \
    \frac{1}{\sqrt{4\pi Dt}}e^{\frac{\left(i(x-x_0)\right)^2}{4Dt}}, \\
    &= \
    \frac{1}{\sqrt{4\pi Dt}}e^{-\frac{(x-x_0)^2}{4Dt}}. \\
\end{align}
$$

This is the same as $\tilde{\phi}(x,t)$ from Ex. 2.1.

#### 2.2.2
We see that $\phi(x,t)$ is a normal distribution in $x$ with
$$
    \mu = x_0 \quad \mathrm{and} \quad \sigma^2 = 2Dt,
$$
where $\mu$ is the expected value, and $\sigma^2$ is the variance. We see immediately that higher diffusion coeffecient means greater variance.

#### 2.2.3
Now let us use the results to find the time evolution $\phi(x,t)$ given the initial condition $\phi(x, 0) = g(x)$, where $g(x)$ is an arbitrary function.

$$
\begin{align}
    \phi(x, t) &= \phi(x, 0) + \
    \sum_{n=1}^{\infty} \frac{t^n}{n!}\left.\frac{\partial^n \phi(x, t')}{\partial t'^n}\right|_{t' = 0}, \\
    &= \
    \phi(x, 0) + \sum_{n=1}^{\infty} \frac{(Dt)^n}{n!}\frac{\partial^{2n} \phi(x, 0)}{\partial x^{2n}}, \\
    &= \
    g(x) + \sum_{n=1}^{\infty} \frac{(Dt)^n}{n!}\frac{\partial^{2n} g(x)}{\partial x^{2n}}, \\
    &= \
    g(x) + \sum_{n=1}^{\infty} \frac{(Dt)^n}{n!}\frac{\partial^{2n}}{\partial x^{2n}}\
    \left( \int_{-\infty}^{\infty}dx_0 \,g(x_0)\,\delta(x-x_0)\right), \\
    &= \
    g(x) + \sum_{n=1}^{\infty} \frac{(Dt)^n}{n!}\frac{\partial^{2n}}{\partial x^{2n}}\
    \left( \int_{-\infty}^{\infty}dx_0 \,g(x_0)\
    \left(\frac{1}{2\pi}\int_{-\infty}^{\infty}dk e^{-ik(x-x_0)}\right)\right), \\
    &= \
    g(x) + \sum_{n=1}^{\infty} \frac{(Dt)^n}{n!}(-ik)^{2n}\
    \left( \int_{-\infty}^{\infty}dx_0 \,g(x_0)\
    \left(\frac{1}{2\pi}\int_{-\infty}^{\infty}dk e^{-ik(x-x_0)}\right)\right), \\
    &= \
    \int_{-\infty}^{\infty}dx_0 \,g(x_0)\
    \left(\frac{1}{2\pi}\int_{-\infty}^{\infty}dk e^{-ik(x-x_0)}\right)\
    + \sum_{n=1}^{\infty} \frac{(-Dtk^2)^n}{n!}\
    \left( \int_{-\infty}^{\infty}dx_0 \,g(x_0)\
    \left(\frac{1}{2\pi}\int_{-\infty}^{\infty}dk e^{-ik(x-x_0)}\right)\right), \\
    &= \
    \sum_{n=0}^{\infty} \frac{(-Dtk^2)^n}{n!}\
    \left( \int_{-\infty}^{\infty}dx_0 \,g(x_0)\
    \left(\frac{1}{2\pi}\int_{-\infty}^{\infty}dk e^{-ik(x-x_0)}\right)\right), \\
    &= \
    \int_{-\infty}^{\infty}dx_0 \,g(x_0)
    \left(\frac{1}{2\pi} \int_{-\infty}^{\infty} dk e^{-Dtk^2}e^{-ik(x-x_0)}\right), \\
    &= \
    \int_{-\infty}^{\infty}dx_0\left(\frac{1}{\sqrt{4\pi Dt}}e^{-\frac{(x-x_0)^2}{4Dt}}\right)g(x_0)\, .
\end{align}
$$

Now we can check the result by setting $f(x_0) = \frac{1}{\sqrt{4\pi Dt}}e^{-\frac{(x-x_0)^2}{4Dt}}$, and $g(x_0) = \delta(x-x_0)$.

$$
    \phi(x,t) = \int_{-\infty}^{\infty}dx_0\, f(x_0)\,\delta(x-x_0) \
    = f(x) \
    = \frac{1}{\sqrt{4\pi Dt}}e^{-\frac{(x-x_0)^2}{4Dt}},
$$
as expected!

## Section 3: Programming - *Random walk in 1D*

```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
import numpy as np
import random
from matplotlib import pyplot as plt
from scipy.stats import norm
import warnings
warnings.filterwarnings("ignore", category=UserWarning, module="matplotlib")

fontsize = 12
newparams = {'axes.titlesize': fontsize, 'axes.labelsize': fontsize,
             'lines.linewidth': 1.5, 'lines.markersize': 12,
             'figure.figsize': (10, 4), 'ytick.labelsize': fontsize,
             'xtick.labelsize': fontsize, 'legend.fontsize': fontsize,
             'legend.handlelength': 1.5, 'figure.autolayout': True}

def randomWalkOneDim():
    N = 1000
    phi = np.zeros(N) # Starting with N=1000 particles at x=0
    deltaT = 1 # steps in time
    steps = 100*deltaT # amount of time steps
    h = 1 # step size
    
    for i in range(steps):
        for j in range(N):
            phi[j] += random.randrange(-h,h+1, 2*h)
    
    # Updating plot parameters
    plt.rcParams.update(newparams)
    width = 6
    height = 4
    plt.figure(figsize=(width, height))
    
    # Fitting a normal distribution to the data
    mu, std = norm.fit(phi)
    
    # Plotting the histogram
    binwidth = 2
    plt.hist(phi, bins=np.arange(min(phi), max(phi) + binwidth, binwidth),
             density=True, alpha=0.75, color='g', ec='k')
    
    # Plotting the PDF
    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 100)
    p = norm.pdf(x, mu, std)
    plt.plot(x, p, 'r--', linewidth=2)
    plt.xlabel(r'$x$')
    plt.ylabel(r'$\phi(x, 100 \Delta t)$')
    title = r"Fit results: $\mu$ = %.2f, $\sigma$ = %.2f" % (mu, std)
    plt.title(title)
    
    plt.show()


randomWalkOneDim()
```

From the histogram above we see that the random walk of 1000 particles starting at $x=0$ spreading out after 100$\Delta t$ time steps is spreading out while being normally distributed.

### Exercise 5.1

```python
k_B = 1.3806503e-23 # [m^2 kg s^{-2} K{-1}] Boltzmann constant
T = 310 # [K] Human body temperature
beta = 1/(k_B * T)

def randomWalkInPot(V, betaK):
    N = 1000
    phi = np.zeros(N) # Starting with N=1000 particles at x=0
    deltaT = 1 # steps in time
    steps = 100*deltaT # amount of time steps
    h = 1 # step size
    k = betaK/beta
    phi = np.zeros(N)
    for i in range(steps):
        if (V == 1):
            DeltaV = V_1(k, phi-h) - V_1(k, phi+h)
        elif (V == 2):
            DeltaV = V_2(k, phi-h, h) - V_2(k, phi+h, h)
        else:
            DeltaV = V_3(k, phi-h, h) - V_3(k, phi+h, h)
        P_minus = 1 - 1/(1 + np.exp(-beta*DeltaV))
        for j in range(N):
            if (random.uniform(0,1) >= P_minus[j]):
                phi[j] += h
            else:
                phi[j] -= h
    return phi

def V_1(k, x):
    return k*x

def V_2(k, x, h):
    n = x.size
    potV = np.zeros(n)
    for i in range(n):
        if (-3*h < x[i] < 3*h):
            potV[i] = k
    return potV

def V_3(k, x, h):
    n = x.size
    potV = np.zeros(n)
    for i in range(n):
        if (x[i] < -3*h):
            potV[i] = -k
        elif (x[i] > 3*h):
            potV[i] = k
        else:
            potV[i] = k*(-1 + 2*(x[i] + 3*h)/(6*h))
    return potV
```

```python
def plot5_1():
    betaK_1 = 0.1
    betaK_2 = 0.7
    betaK_3 = 2.2
    betaK_4 = 5.0
    phi_1 = randomWalkInPot(1, betaK_1)
    phi_2 = randomWalkInPot(1, betaK_2)
    phi_3 = randomWalkInPot(1, betaK_3)
    phi_4 = randomWalkInPot(1, betaK_4)
    
    # Set up plot
    width = 9
    height = 5
    plt.figure(figsize=(width, height))
    binwidth = 2
    binsArr = np.arange(min(phi_4), max(phi_1) + binwidth, binwidth)
    
    #Plot
    plt.subplot(2, 2, 1)
    plt.hist(phi_1, bins=binsArr, alpha=0.75, color='tab:orange', ec='k', label=r'$N(x)$')
    plt.title(r"$\beta k = $ %.1f" %betaK_1)
    plt.xlabel(r'$x$')
    plt.legend(loc='best')
    
    plt.subplot(2, 2, 2)
    plt.hist(phi_2, bins=binsArr, alpha=0.75, color='tab:orange', ec='k', label=r'$N(x)$')
    plt.title(r"$\beta k = $ %.1f" %betaK_2)
    plt.xlabel(r'$x$')
    plt.legend(loc='best')
    
    plt.subplot(2, 2, 3)
    plt.hist(phi_3, bins=binsArr, alpha=0.75, color='tab:orange', ec='k', label=r'$N(x)$')
    plt.title(r"$\beta k = $ %.1f" %betaK_3)
    plt.xlabel(r'$x$')
    plt.legend(loc='best')
    
    plt.subplot(2, 2, 4)
    plt.hist(phi_4, bins=binsArr, alpha=0.75, color='tab:orange', ec='k', label=r'$N(x)$')
    plt.title(r"$\beta k = $ %.1f" %betaK_4)
    plt.xlabel(r'$x$')
    plt.legend(loc='best')

plot5_1()
```

## 7 Programming *Interplay between diffusion and voltage*
### 7.1

```python
e = 1.6021766e-19 # [C] elementary charge

def diffAndVoltage(betaU_Na, betaU_K):
    L = 50 # Length of system along x-axis
    U0_Na = betaU_Na/beta # Potential for sodium and potassium channels
    U0_K = betaU_K/beta
    h = 1 # half the membrane depht
    DeltaT = 1 # time step
    steps = 500 # time steps
    
    V = np.zeros(steps) # the voltage
    
    Na_int = -L/4 * np.ones(60) # interior
    Na_ext = L/4 * np.ones(1450) # exterior
    phi_Na = np.concatenate((Na_int, Na_ext)) # initial am. of ions along axis
    sizeNa = phi_Na.size 
    
    K_int = -L/4 * np.ones(1400) # interior
    K_ext = L/4 * np.ones(50) # exterior
    phi_K = np.concatenate((K_int, K_ext)) # initial amount of ions along axis
    sizeK = phi_K.size
    
    for i in range(steps):
        V[i] = Volt(phi_Na, phi_K, h, L)
        DeltaU_Na = U(V[i], phi_Na - h, U0_Na, h) - U(V[i], phi_Na + h, U0_Na, h)
        DeltaU_K = U(V[i], phi_K - h, U0_K, h) - U(V[i], phi_K + h, U0_K, h)
        
        P_minus_Na = 1 - 1/(1 + np.exp(-beta*DeltaU_Na))
        P_minus_K = 1 - 1/(1 + np.exp(-beta*DeltaU_K))
        
        for j in range(sizeNa):
            if phi_Na[j] == -L/2 + 0.5:
                phi_Na[j] += h
            elif phi_Na[j] == L/2 - 0.5:
                phi_Na[j] -= h
            else:
                if (random.uniform(0,1) >= P_minus_Na[j]):
                    phi_Na[j] += h
                else:
                    phi_Na[j] -= h
                    
        for j in range(sizeK):
            if phi_K[j] == -L/2 + 0.5:
                phi_K[j] += h
            elif phi_K[j] == L/2 - 0.5:
                phi_K[j] -= h
            else:
                if (random.uniform(0,1) >= P_minus_K[j]):
                    phi_K[j] += h
                else:
                    phi_K[j] -= h
    return 1e3*V, phi_Na, phi_K

def U(Vt, phi, U0, h):
    n = phi.size
    u = np.zeros(n)
    for i in range(n):
        if (phi[i] < -h):
            u[i] = e*Vt
        elif (-h <= phi[i] <= h):
            u[i] = -e*Vt*(phi[i]-h)/(2*h) + U0
    return u

def Volt(na, k, h, L):
    Cc = 0.07*e # [e M/V] Concentration capacitance
    partCon = 0.1 # concentration per particle
    extCon = (1450 + 50) * partCon
    intCon = partCon*(((-L/2 <= na) & (na < 0)).sum()\
                      + ((-L/2 <= k) & (k < 0)).sum())
    return 1e-3*(intCon - extCon)*e / Cc # multiplying with 10^-3, from mV to V
```

```python
def plot7_1():
    betaU_1 = 0
    betaU_2 = 1
    betaU_3 = 3
    V_1, Na_1, K_1 = diffAndVoltage(betaU_1, betaU_1)
    V_2, Na_2, K_2 = diffAndVoltage(betaU_2, betaU_2)
    V_3, Na_3, K_3 = diffAndVoltage(betaU_3, betaU_3)
    
    # Set up plot
    width = 10
    height = 8
    plt.figure(figsize=(width, height))
    binwidth = 2
    binsNa = np.arange(min(np.concatenate((Na_1, Na_2, Na_3))),
                       max(np.concatenate((Na_1, Na_2, Na_3))) + binwidth, binwidth)
    binsK = np.arange(min(np.concatenate((K_1, K_2, K_3))),
                      max(np.concatenate((K_1, K_2, K_3))) + binwidth, binwidth)
    
    t = np.arange(V_1.size)
    
    #Plot
    plt.subplot(3, 3, 1)
    plt.plot(t, V_1, 'tab:red', label=r'Voltage $V(t)$')
    plt.title(r"$\beta U_{\mathrm{Na}}^0 = \beta U_{\mathrm{K}}^0 = $ %.f" %betaU_1)
    plt.xlabel(r'$t$')
    plt.legend(loc='best')
    
    plt.subplot(3, 3, 2)
    plt.plot(t, V_2, 'tab:red', label=r'Voltage $V(t)$')
    plt.title(r"$\beta U_{\mathrm{Na}}^0 = \beta U_{\mathrm{K}}^0 = $ %.f" %betaU_2)
    plt.xlabel(r'$t$')
    plt.legend(loc='best')
    
    plt.subplot(3, 3, 3)
    plt.plot(t, V_3, 'tab:red', label=r'Voltage $V(t)$')
    plt.title(r"$\beta U_{\mathrm{Na}}^0 = \beta U_{\mathrm{K}}^0 = $ %.f" %betaU_3)
    plt.xlabel(r'$t$')
    plt.legend(loc='best')
    
    plt.subplot(3, 3, 4)
    plt.hist(Na_1, bins=binsNa, alpha=0.75, color='tab:orange', ec='k', label=r'$N_\mathrm{Na}$')
    plt.title(r"$\beta U_{\mathrm{Na}}^0 = $ %.f" %betaU_1)
    plt.xlabel(r'$x$')
    plt.legend(loc='best')
    
    plt.subplot(3, 3, 5)
    plt.hist(Na_2, bins=binsNa, alpha=0.75, color='tab:orange', ec='k', label=r'$N_\mathrm{Na}$')
    plt.title(r"$\beta U_{\mathrm{Na}}^0 = $ %.f" %betaU_2)
    plt.xlabel(r'$x$')
    plt.legend(loc='best')
    
    plt.subplot(3, 3, 6)
    plt.hist(Na_3, bins=binsNa, alpha=0.75, color='tab:orange', ec='k', label=r'$N_\mathrm{Na}$')
    plt.title(r"$\beta U_{\mathrm{Na}}^0 = $ %.f" %betaU_3)
    plt.xlabel(r'$x$')
    plt.legend(loc='best')
    
    plt.subplot(3, 3, 7)
    plt.hist(K_1, bins=binsK, alpha=0.75, color='tab:purple', ec='k', label=r'$N_\mathrm{K}$')
    plt.title(r"$\beta U_{\mathrm{K}}^0 = $ %.f" %betaU_1)
    plt.xlabel(r'$x$')
    plt.legend(loc='best')
    
    plt.subplot(3, 3, 8)
    plt.hist(K_1, bins=binsK, alpha=0.75, color='tab:purple', ec='k', label=r'$N_\mathrm{K}$')
    plt.title(r"$\beta U_{\mathrm{K}}^0 = $ %.f" %betaU_2)
    plt.xlabel(r'$x$')
    plt.legend(loc='best')
    
    plt.subplot(3, 3, 9)
    plt.hist(K_1, bins=binsK, alpha=0.75, color='tab:purple', ec='k', label=r'$N_\mathrm{K}$')
    plt.title(r"$\beta U_{\mathrm{K}}^0 = $ %.f" %betaU_3)
    plt.xlabel(r'$x$')
    plt.legend(loc='best')
    
    plt.show()

plot7_1()
```
